```python
import datetime, os, tweepy
from twitterbox import load_api, StreamListener
```

```python
project = 'test'
folder = 'data'
file = datetime.datetime.now().strftime("%Y%m%d")
file = os.path.join(folder, f'{project}_{file}.csv')
query = ['humala', 'keiko', 'fujimori']
lang = ['es']
tweet_limit = 50
```

```python
api = load_api()
listener = StreamListener(file, lang, query, tweet_limit)
myStream = tweepy.Stream(auth=api.auth, listener=listener)

myStream.filter(follow=None,
                track=query,
                locations=None,
                languages=lang,
                async=False)
```

```python
myStream.disconnect()
myStream.running
```
